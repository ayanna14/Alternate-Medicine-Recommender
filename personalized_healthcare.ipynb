```python
data['Description'] = data['Description'].apply(lambda x:x.split())
data['Reason'] = data['Reason'].apply(lambda x:x.split())
```

```python
data.tail()
```

```python
data['Description'] = data['Description'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
data['tags'] = data['Description'] + data['Reason']
```

```python
data.head()
```

```python
updated_data = data[['index','Drug_Name','tags']]
```

```python
updated_data.head()
```

```python
updated_data['tags'] = updated_data['tags'].apply(lambda x:" ".join(x))
```

```python
updated_data['tags'] = updated_data['tags'].apply(lambda x:x.lower())
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(stop_words = "english", max_features = 5000)
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
stem("Hello Python Pandas")
```

```python
updated_data['tags'] = updated_data['tags'].apply(stem)
```

```python
updated_data.head()
```

```python
vectors = cv.fit_transform(updated_data['tags']).toarray()
```

```python
vectors.shape
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
def recommend(medicine):
    medicine_index = updated_data[updated_data['Drug_Name'] == medicine].index[0]
    distance = similarity[medicine_index]
    medicine_list = sorted(list(enumerate(distance)), reverse=True, key = lambda x : x[1])[1:6]
    for i in medicine_list:
        print(updated_data.iloc[i[0]].Drug_Name)
```

```python
#test case 1
recommend("ACGEL CL NANO Gel 15gm")
```

```python
#test case 2
recommend("Dmentin MD 10mg Tablet 10'SDmentin MD 5mg Tablet 10'S")
```
